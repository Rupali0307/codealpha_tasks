```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('/content/unemployment.csv.csv')

# Clean column names
df.columns = df.columns.str.strip()

# Check columns
print(df.columns)
df.head()
```

```python
# Convert to datetime if 'Date' column exists
df['Date'] = pd.to_datetime(df['Date'])

# Extract year if needed
df['Year'] = df['Date'].dt.year
```

```python
# Line plot over time
plt.figure(figsize=(12,6))
sns.lineplot(x='Date', y='Estimated Unemployment Rate (%)', data=df)
plt.title('Unemployment Rate Over Time')
plt.xticks(rotation=45)
plt.ylabel('Unemployment Rate (%)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(14,6))
sns.barplot(x='Region', y='Estimated Unemployment Rate (%)', data=df)
plt.xticks(rotation=90)
plt.title('Region-wise Unemployment Rate')
plt.show()
```

```python
pre_covid = df[df['Year'] < 2020]['Estimated Unemployment Rate (%)'].mean()
post_covid = df[df['Year'] >= 2020]['Estimated Unemployment Rate (%)'].mean()

print("Average Unemployment Before COVID:", round(pre_covid, 2), "%")
print("Average Unemployment After COVID:", round(post_covid, 2), "%")
```
